```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('heap_a.csv')
df2 = pd.read_csv('quick_a.csv')
df3 = pd.read_csv('insertion_a.csv')
df4 = pd.read_csv('merge_a.csv')
df5 = pd.read_csv('bubble_a.csv')

bysize1 = [df1.loc[df1["size"] == 100],
          df1.loc[df1["size"] == 1000],
          df1.loc[df1["size"] == 10000]]
bysize2 = [df2.loc[df2["size"] == 100],
          df2.loc[df2["size"] == 1000],
          df2.loc[df2["size"] == 10000]]        
bysize3 = [df3.loc[df3["size"] == 100],
          df3.loc[df3["size"] == 1000],
        df3.loc[df3["size"] == 10000]]
bysize4 = [df4.loc[df4["size"] == 100],
          df4.loc[df4["size"] == 1000],
        df4.loc[df4["size"] == 10000]]
bysize5 = [df5.loc[df5["size"] == 100],
          df5.loc[df5["size"] == 1000],
        df5.loc[df5["size"] == 10000]]
```

```python
x1 = np.array(df1['size'].unique())
x2 = np.array(df2['size'].unique())
x3 = np.array(df3['size'].unique())
x4 = np.array(df4['size'].unique())
x5 = np.array(df5['size'].unique())

y1 = [np.mean(x1['time']) for x1 in bysize1]
y2 = [np.mean(x2['time']) for x2 in bysize2]
y3 = [np.mean(x3['time']) for x3 in bysize3]
y4 = [np.mean(x4['time']) for x4 in bysize4]
y5 = [np.mean(x5['time']) for x5 in bysize5]
```

```python
plt.figure(figsize=[10, 10], facecolor='grey')
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio [s]')

plt.plot(x1,y1, 'o--', label='heap')
plt.plot(x2,y2, 'o--', label='quick')
""" plt.plot(x3,y3, 'o--', label='insertion') """
plt.plot(x4,y4, 'o--', label='merge')
""" plt.plot(x5,y5, 'o--', label='bubble') """

plt.legend()
```
