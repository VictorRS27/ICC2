```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('bucket_a.csv')
df2 = pd.read_csv('counting_a.csv')
df3 = pd.read_csv('radix_a.csv')

bysize1 = [df1.loc[df1["size"] == 100],
          df1.loc[df1["size"] == 500],
          df1.loc[df1["size"] == 1000],
          df1.loc[df1["size"] == 2500],
          df1.loc[df1["size"] == 5000],
          df1.loc[df1["size"] == 7500],
          df1.loc[df1["size"] == 10000],
          df1.loc[df1["size"] == 25000],
          df1.loc[df1["size"] == 50000],
          df1.loc[df1["size"] == 75000],
          df1.loc[df1["size"] == 100000]]
bysize2 = [df2.loc[df2["size"] == 100],
          df2.loc[df2["size"] == 500],
          df2.loc[df2["size"] == 1000],
          df2.loc[df2["size"] == 2500],
          df2.loc[df2["size"] == 5000],
          df2.loc[df2["size"] == 7500],
          df2.loc[df2["size"] == 10000],
          df2.loc[df2["size"] == 25000],
          df2.loc[df2["size"] == 50000],
          df2.loc[df2["size"] == 75000],
          df2.loc[df2["size"] == 100000]]
bysize3 = [df3.loc[df3["size"] == 100],
          df3.loc[df3["size"] == 500],
          df3.loc[df3["size"] == 1000],
          df3.loc[df3["size"] == 2500],
          df3.loc[df3["size"] == 5000],
          df3.loc[df3["size"] == 7500],
          df3.loc[df3["size"] == 10000],
          df3.loc[df3["size"] == 25000],
          df3.loc[df3["size"] == 50000],
          df3.loc[df3["size"] == 75000],
          df3.loc[df3["size"] == 100000]]
```

```python
x1 = np.array(df1['size'].unique())
x2 = np.array(df2['size'].unique())
x3 = np.array(df3['size'].unique())

y1 = [np.mean(x1['time']) for x1 in bysize1]
y2 = [np.mean(x2['time']) for x2 in bysize2]
y3 = [np.mean(x3['time']) for x3 in bysize3]
```

```python
print(y1)
```

```python
plt.figure(figsize=[10, 10], facecolor='grey')
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio [s]')

plt.plot(x1,y1, 'o--', label='bucket')
plt.plot(x2,y2, 'o--', label='counting')
plt.plot(x3,y3, 'o--', label='radix')

plt.legend()
```
